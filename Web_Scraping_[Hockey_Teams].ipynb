```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.scrapethissite.com/pages/forms/"
```

```python
#get html content
content = requests.get(url)
htmlContent = content.content
print(htmlContent)
```

```python
#parse the HTML
Soup = BeautifulSoup(htmlContent)
print(Soup.prettify())
```

```python
cards = Soup.find_all("td", class_="name")
name = []
for card in cards:
    name.append(card.text)  # Access the text of each individual div element
print(len(name))
print(name)
```

```python
cards = Soup.find_all("td", class_="year")
year = []
for card in cards:
    year.append(card.text)  # Access the text of each individual div element
print(len(year))
print(year)
```

```python
cards = Soup.find_all("td", class_="wins")
wins = []
for card in cards:
    wins.append(card.text)  # Access the text of each individual div element
print(len(wins))
print(wins)
```

```python
cards = Soup.find_all("td", class_="losses")
losses = []
for card in cards:
    losses.append(card.text)  # Access the text of each individual div element
print(len(losses))
print(losses)
```

```python
cards = Soup.find_all("td", class_="pct")
pct = []
for card in cards:
    pct.append(card.text)  # Access the text of each individual div element
print(len(pct))
print(pct)
```

```python
cards = Soup.find_all("td", class_="gf")
goalsfor = []
for card in cards:
    goalsfor.append(card.text)  # Access the text of each individual div element
print(len(goalsfor))
print(goalsfor)
```

```python
cards = Soup.find_all("td", class_="ga")
goalsagainst = []
for card in cards:
    goalsagainst.append(card.text)  # Access the text of each individual div element
print(len(goalsagainst))
print(goalsagainst)
```

```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
df = pd.DataFrame({"Team Name":name,"Year":year,"Wins":wins,"Losses":losses,"Win %":pct,"Goals For":goalsfor,"Goals Against":goalsagainst})
print(df)
```

```python
df["Team Name"].iloc[:] = df["Team Name"].iloc[:].str.strip()
df.Year.iloc[:] = df.Year.iloc[:].str.strip()
df.Wins.iloc[:] = df.Wins.iloc[:].str.strip()
df.Losses.iloc[:] = df.Losses.iloc[:].str.strip()
df["Win %"].iloc[:] = df["Win %"].iloc[:].str.strip()
df["Goals For"].iloc[:] = df["Goals For"].iloc[:].str.strip()
df["Goals Against"].iloc[:] = df["Goals Against"].iloc[:].str.strip()
```

```python
df["Year"] = df["Year"].astype(int)
df["Wins"] = df["Wins"].astype(int)
df["Losses"] = df["Losses"].astype(int)
df["Goals For"] = df["Goals For"].astype(int)
df["Goals Against"] = df["Goals Against"].astype(int)
df['Win %'] = df['Win %'].astype(float)
```

```python
#List of all Teams
df["Team Name"].unique()
```

```python
a = df[df["Year"]==1991]
a
```

```python
#pie-chart showing wins of the teams played in year 1991
fig = px.pie(a,values="Wins",names="Team Name",hover_data = ["Goals For"],title = "Win Percentage of the teams played in 1991")
fig.update_layout(uniformtext_minsize=20, uniformtext_mode='hide', title_font=dict(size=40)) # Add this line
fig.show()
```

```python
#Showing the records of the teams with maximum goals
b = df.sort_values(by = "Goals For",ascending = False)[["Team Name","Year","Win %","Goals For"]].head(5)
b
```

```python
fig = px.bar(b,x="Team Name",y="Goals For",title = "Most Goals by a Team",color = "Team Name")
fig.update_layout(title_font=dict(size=40))
fig.show()
```

```python
fig = px.scatter(df,x="Team Name",y="Goals For",title = "Total Goals scored by Teams in 1990 AND 1991",size = "Goals For",color = "Year")
fig.update_layout(title_font=dict(size=40)) # Add this line
fig.show()
```
