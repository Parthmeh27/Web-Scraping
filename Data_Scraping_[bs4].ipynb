```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.bookswagon.com/"
```

```python
#get html content
content = requests.get(url)
htmlContent = content.content
print(htmlContent)
```

```python
#parse the HTML
Soup = BeautifulSoup(htmlContent)
print(Soup.prettify())
```

```python
container = Soup.find_all("span", class_="booktitle")
title = []
for card in container:
    title.append(card.text)  # Access the text of each individual div element
print(title)
```

```python
container = Soup.find_all("span", class_="author")
author_name = []
for div in container:
    author_name.append(div.text)  # Access the text of each individual div element
```

```python
container = Soup.find_all("input", class_="avergageratingslider")
ratings = []
for div in container:
    value = div.get('value')
    ratings.append(value)
```

```python
container = Soup.find_all("span", class_="actualprice")
sp = []
for div in container:
    sp.append(div.text)  # Access the text of each individual div element
sp.pop()
```

```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
df = pd.DataFrame({'Title': title, 'Author': author_name, 'Ratings': ratings, 'Price': sp})
print(df)
```

```python
df.Title.iloc[:] = df.Title.iloc[:].str.strip()
df.Author.iloc[:] = df.Author.iloc[:].str.strip()
```

```python
df["Price"].replace("", "", inplace=True)
df["Price"].replace(",", "", inplace=True)
df["Price"] = df["Price"].astype(int)
df
```
